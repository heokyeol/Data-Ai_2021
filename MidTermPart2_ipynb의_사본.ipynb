````
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
                                                                    
# Question 3: Explain how the categories of the images are extracted when you execute the above statement.
# Specify the proportion of the factor valid_pct in the function as valid dataset. In other words, in
# the above function, 20% of the pictures in the folder are valid dataset.
````
